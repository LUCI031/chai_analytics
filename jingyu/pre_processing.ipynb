# 데이터 전처리

```python
# 시계열 처리

df['created_at'] = pd.to_datetime(df['created_at'])
df['birthday'] = pd.to_datetime(df['birthday'])
df['sign_up_date'] = pd.to_datetime(df['sign_up_date'])
```

```python
# gender

df.loc[df['gender']=='male','gender'] = 0
df.loc[df['gender']=='female','gender'] = 1
```
